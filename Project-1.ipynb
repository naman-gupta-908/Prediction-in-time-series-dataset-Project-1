Time series models have always been of utmost importance. In simple words, time series analysis allows us to analyze past events and help us make predictions for the future. Organizations, therefore, rely on time series analysis to make better business decisions. With this, they also have the ability to stand ahead in the race. Although there are numerous python libraries for time series analysis here the work will be done with Prophet an one open-source python time-series library.

The prophet is one open-source python time-series library that is dedicated to making predictions for one-dimensional time-series datasets. Its capability is such that it can make accurate predictions for data with the trend and seasonal structure by default. Also, it is quite user-friendly.

```python
op=pd.DataFrame(columns=['sec_no','parameter','Actual_Value','Predicted_Value'])
op_rmse=pd.DataFrame(columns=['sec_no','RMSE'])
```

forcasting for each section

```python
with tf.device('/GPU:0'):
  begin=0
  end=10
  while end!=1020:
    df_tmp=df_new.iloc[begin:end,:]
    # print(df_tmp)
    scaler = StandardScaler()
    df_tmp.loc[:,1:] = scaler.fit_transform(df_tmp.iloc[:,1:])
    df_tmp.rename(columns={'year':'ds','Para-9':'y','Para-10':'y','Para-11':'y','Para-12':'y','Para-13':'y'},inplace=True)
    
    # Model for Para-9
    df_tmp9=pd.concat([df_tmp.iloc[:9,:10],df_tmp.iloc[:9,10]],axis=1)

    m9=Prophet(interval_width=0.90,seasonality_mode='multiplicative')
    m9.add_regressor('Para-1')
    m9.add_regressor('Para-2')
    m9.add_regressor('Para-3')
    m9.add_regressor('Para-4')
    m9.add_regressor('Para-5')
    m9.add_regressor('Para-6')
    m9.add_regressor('Para-7')
    m9.add_regressor('Para-8')
    # print(df_tmp9)
    m9.fit(df_tmp9)
    op9=df_tmp.iloc[8:10,:10]
    pred9=m9.predict(op9)


    # Model for Para-10
    df_tmp10=pd.concat([df_tmp.iloc[:9,:10],df_tmp.iloc[:9,11]],axis=1)

    m10=Prophet(interval_width=0.90,seasonality_mode='multiplicative')
    m10.add_regressor('Para-1')
    m10.add_regressor('Para-2')
    m10.add_regressor('Para-3')
    m10.add_regressor('Para-4')
    m10.add_regressor('Para-5')
    m10.add_regressor('Para-6')
    m10.add_regressor('Para-7')
    m10.add_regressor('Para-8')
    m10.fit(df_tmp10)
    op10=df_tmp.iloc[8:10,:10]
    pred10=m10.predict(op10)

    
    # Model for Para-11
    df_tmp11=pd.concat([df_tmp.iloc[:9,:10],df_tmp.iloc[:9,12]],axis=1)

    m11=Prophet(interval_width=0.90,seasonality_mode='multiplicative')
    m11.add_regressor('Para-1')
    m11.add_regressor('Para-2')
    m11.add_regressor('Para-3')
    m11.add_regressor('Para-4')
    m11.add_regressor('Para-5')
    m11.add_regressor('Para-6')
    m11.add_regressor('Para-7')
    m11.add_regressor('Para-8')
    m11.fit(df_tmp11)
    op11=df_tmp.iloc[8:10,:10]
    pred11=m11.predict(op11)

    
    # Model for Para-12
    df_tmp12=pd.concat([df_tmp.iloc[:9,:10],df_tmp.iloc[:9,13]],axis=1)

    m12=Prophet(interval_width=0.90,seasonality_mode='multiplicative')
    m12.add_regressor('Para-1')
    m12.add_regressor('Para-2')
    m12.add_regressor('Para-3')
    m12.add_regressor('Para-4')
    m12.add_regressor('Para-5')
    m12.add_regressor('Para-6')
    m12.add_regressor('Para-7')
    m12.add_regressor('Para-8')
    m12.fit(df_tmp12)
    op12=df_tmp.iloc[8:10,:10]
    pred12=m12.predict(op12)

        
    # Model for Para-13
    df_tmp13=pd.concat([df_tmp.iloc[:9,:10],df_tmp.iloc[:9,14]],axis=1)

    m13=Prophet(interval_width=0.90,seasonality_mode='multiplicative')
    m13.add_regressor('Para-1')
    m13.add_regressor('Para-2')
    m13.add_regressor('Para-3')
    m13.add_regressor('Para-4')
    m13.add_regressor('Para-5')
    m13.add_regressor('Para-6')
    m13.add_regressor('Para-7')
    m13.add_regressor('Para-8')
    m13.fit(df_tmp13)
    op13=df_tmp.iloc[8:10,:10]
    pred13=m13.predict(op13)

    # op
    actual=[]
    predicted=[]
    actual.append(df_tmp.iloc[9,9])
    actual.append(df_tmp.iloc[9,10])
    actual.append(df_tmp.iloc[9,11])
    actual.append(df_tmp.iloc[9,12])
    actual.append(df_tmp.iloc[9,13])

    predicted.append(pred9['yhat'][1])
    predicted.append(pred10['yhat'][1])
    predicted.append(pred11['yhat'][1])
    predicted.append(pred12['yhat'][1])
    predicted.append(pred13['yhat'][1])

    op.loc[len(op)]=[end//10,'Para9',df_tmp.iloc[9,9],pred9['yhat'][1]]
    op.loc[len(op)]=[end//10,'Para10',df_tmp.iloc[9,10],pred10['yhat'][1]]
    op.loc[len(op)]=[end//10,'Para11',df_tmp.iloc[9,11],pred11['yhat'][1]]
    op.loc[len(op)]=[end//10,'Para12',df_tmp.iloc[9,12],pred12['yhat'][1]]
    op.loc[len(op)]=[end//10,'Para13',df_tmp.iloc[9,13],pred13['yhat'][1]]
    tmp = np.sqrt(mean_squared_error(actual, predicted))
    op_rmse.loc[len(op_rmse)]=[int(end//10),tmp]
    begin+=10
    end+=10
```
